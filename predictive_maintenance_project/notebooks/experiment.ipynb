## Feature Engineering

```python
#Converting Celcius to Kelvin, 1 K= -272 C
df["Air temperature [C]"]=df["Air temperature [K]"]-272
```

```python
df.head()
```

```python
df["Process temperature [C]"]=df["Process temperature [K]"]-272
```

```python
df.head()
```

```python
df["Temperature Difference[C]"]=df["Process temperature [K]"]-df["Air temperature [K]"]
```

```python
df.head()
```

Power Output Formula:
Pout=TW,
w=2x3.14xr/60
Pout=Output Power
T=Torque
w=Speed Velocity
r=rotational speed

```python
df["Output Power"]=df["Torque [Nm]"]*((2*3.14*df["Rotational speed [rpm]"])/60)
```

```python
df.head()
```

```python
df["Output Power (Kw)"]=df["Output Power"]/1000
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df["Type_Encoded"]=le.fit_transform(df["Type"])
```

```python
df.head()
```

```python
df["FailureType_Encoded"]=le.fit_transform(df["Failure Type"])
```

```python
df=df.drop(["Type","Air temperature [K]","Process temperature [K]","Failure Type","Output Power"],axis=1)
```
