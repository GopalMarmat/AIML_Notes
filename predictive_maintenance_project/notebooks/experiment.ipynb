## Feature Engineering

```python
#Converting Celcius to Kelvin, 1 K= -272 C
df["Air temperature [C]"]=df["Air temperature [K]"]-272
```

```python
df.head()
```

```python
df["Process temperature [C]"]=df["Process temperature [K]"]-272
```

```python
df.head()
```

```python
df["Temperature Difference[C]"]=df["Process temperature [K]"]-df["Air temperature [K]"]
```

```python
df.head()
```

Power Output Formula:
Pout=TW,
w=2x3.14xr/60
Pout=Output Power
T=Torque
w=Speed Velocity
r=rotational speed

```python
df["Output Power"]=df["Torque [Nm]"]*((2*3.14*df["Rotational speed [rpm]"])/60)
```

```python
df.head()
```

```python
df["Output Power (Kw)"]=df["Output Power"]/1000
```

```python
pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df["Type_Encoded"]=le.fit_transform(df["Type"])
```

```python
df.head()
```

```python
df["FailureType_Encoded"]=le.fit_transform(df["Failure Type"])
```

```python
df=df.drop(["Type","Air temperature [K]","Process temperature [K]","Failure Type","Output Power"],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["FailureType_Encoded","Type_Encoded"],axis=1)
y=df["FailureType_Encoded"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
pip install mlflow
```

```python
import mlflow
import mlflow.sklearn
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
```

```python
models={
    "lr":LogisticRegression(max_iter=100),
    "dtc":DecisionTreeClassifier(),
    "rfc":RandomForestClassifier(),
    "gbc":GradientBoostingClassifier(),
    "abc":AdaBoostClassifier(),
    "svc":SVC(),
    "knn":KNeighborsClassifier()
}
```

```python
mlflow.set_experiment("predictive_maintenance_classification")
for name,model in models.items():
    with mlflow.start_run(run_name=name):
        #model training
        model.fit(x_train,y_train)
        #predictions
        y_pred=model.predict(x_test)
        #Evalution
        acc=accuracy_score(y_test,y_pred)
        f1=f1_score(y_test,y_pred,average='micro')
        recall=recall_score(y_test,y_pred,average='micro')
        precision=precision_score(y_test,y_pred,average='micro')
        print(model)
        print("accuracy_score",acc)
        print("f1_score",f1)
        print("recall_score",recall)
        print("precision_score",precision)
        print("----------------------------------------------")

        #experment tracking
        mlflow.log_metric("accuracy",acc)
        mlflow.log_metric("f1",f1)
        mlflow.log_metric("recall",recall)
        mlflow.log_metric("precision",precision)
```

## HyperParameter Tunning
### Hyperparameter Tunning is techniques which used to finding best model parameter, to achieve better performance, GridSearchCV and RandomSearch these methods we used in Hyperparameter tunning.

```python
#Grid Search CV
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid={
    'n_estimators':[100,200],
    'max_depth':[5,10,None],
    'min_samples_split':[2,5],
    'min_samples_leaf':[1,2],
    'bootstrap':[True,False]
}

grid_search=GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    n_jobs=1,
    verbose=1
)
grid_search.fit(x_train,y_train)
print("Model Best Parameter:",grid_search.best_params_)
```

```python
import numpy as np
param_random={
    'n_estimators':np.arange(50,200,100),
    'max_depth':[None]+list(np.arange(5,30,5)),
    'min_samples_split':np.arange(2,10),
    'min_samples_leaf':np.arange(1,10),
    'bootstrap':[True,False]
}

random_search=RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_random,
    cv=5,
    n_jobs=1,
    verbose=1,
)
random_search.fit(x_train,y_train)
print("Random Search Model Best Parameter",random_search.best_params_)
```
