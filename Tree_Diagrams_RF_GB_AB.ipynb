# ðŸŒ² Tree Diagrams: Random Forest, Gradient Boosting, AdaBoost

This notebook visualizes the first decision tree from:
- Random Forest
- Gradient Boosting
- AdaBoost

---

```python
# Importing necessary libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## ðŸŒ³ Random Forest - Tree Diagram

```python
# Train Random Forest
rf = RandomForestClassifier(n_estimators=10, random_state=42)
rf.fit(X_train, y_train)

# Plot first tree
plt.figure(figsize=(12, 6))
plot_tree(rf.estimators_[0], filled=True, feature_names=[f"Feature {i}" for i in range(X.shape[1])])
plt.title("Random Forest - Tree 1")
plt.show()
```

## ðŸ” AdaBoost - Tree Diagram

```python
# Train AdaBoost
ab = AdaBoostClassifier(n_estimators=10, random_state=42)
ab.fit(X_train, y_train)

# Plot first weak learner
plt.figure(figsize=(12, 6))
plot_tree(ab.estimators_[0], filled=True, feature_names=[f"Feature {i}" for i in range(X.shape[1])])
plt.title("AdaBoost - Tree 1")
plt.show()
```

## ðŸŽ¯ Gradient Boosting - Tree Diagram

```python
# Train Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=10, random_state=42)
gb.fit(X_train, y_train)

# Plot first weak learner
plt.figure(figsize=(12, 6))
plot_tree(gb.estimators_[0, 0], filled=True, feature_names=[f"Feature {i}" for i in range(X.shape[1])])
plt.title("Gradient Boosting - Tree 1")
plt.show()
```
